```python
def predict(X_new):
    # Load the trained model from the pickled file
    with open("rf_model.pkl", "rb") as f:
        rf = pickle.load(f)

    # Use the model to make predictions on the new data
    y_pred = rf.predict(X_new)

    # Return the predicted value
    return y_pred[0]
```
